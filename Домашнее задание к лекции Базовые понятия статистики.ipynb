```python
# Посчитаем выбросы surgery
q1 = df.surgery.quantile(0.25)
q3 = df.surgery.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.surgery.between(lower_bound, upper_bound, inclusive=True)].sort_values('surgery')
remove_outliers
```

```python
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

```python
for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# surgery - логично было бы заполнить на основе столбца с хирургическим вмешательством

df.surgery.fillna(df.groupby('surgical lesion').surgery.transform('median'), inplace=True)
```

```python
# rectal temperature - мне кажется здесь будет верней рассчитать среднюю температуру
# каждая лошадь индивидуальна и четкой зависимсти температуры от чего-либо в конкретный момент времени мы не знаем

df['rectal temperature'].fillna(df['rectal temperature'].mean(), inplace=True)
```

mucous membranes, capillary refill time - закономерностей не нашел, так как это категорийный вид, то можно не заполнять.

total protein, pulse  - индивидуальный дискретный параметр. Я бы поставил среднее значение или на основе двух и более групп, посоветовавшись со специалистом по лошадям

```python
df['total protein'].fillna(df['total protein'].mean(), inplace=True)
df.pulse.fillna(df.pulse.mean(), inplace=True)

# или пример с заменой по столбцам

df['total protein'].fillna(df.groupby(['pulse', 'peripheral pulse'])['total protein'].transform('median'), inplace=True)
```
