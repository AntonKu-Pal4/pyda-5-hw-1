# Задание 1

```python
import requests
from bs4 import BeautifulSoup
```

```python
KEYWORDS = ['python', 'knative']
```

```python
req = requests.get('https://habr.com/ru/all/all/page4/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
posts = soup.find_all('div', class_='post__text')
posts
```

```python
posts = soup.find_all('article', class_='post')
posts
```

```python
import pandas as pd
```

```python
kom_news = pd.DataFrame()
for post in posts:
    post_lower = post.text.lower() 

    if any([word in post_lower for word in KEYWORDS]):

        date = post.find('span', class_='post__time').text

        title = post.find('a', class_='post__title_link').text

        title_url = post.find('a', class_='post__title_link').attrs.get('href')
        
        row = {'date': date, 'title': title, 'title_url': title_url}
        kom_news = pd.concat([kom_news, pd.DataFrame([row])])  
kom_news.head()
           
```
